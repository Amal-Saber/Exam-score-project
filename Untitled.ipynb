```python
# I need two columns, the total score for the student and the
# percentage for the student, so I added them

df['grade']=((df['math score']+df['reading score']+df['writing score'])/300)*100
df['Total scores']=df['math score']+df['reading score']+df['writing score']

# I will convert the Student Percentage column to Grade
# If a percentage is higher than 80 percent, it will be graded excellent
# Less than 80 and above 60 is a good grade
# Less than 60 is a bad grade

i=0
for p in df['grade'] :
    if p >= 80 :
        df.iloc[i,8]= "excellent"
        
    elif (p < 80) & (p > 60):
        df.iloc[i,8]='good'
        
    else:  
        df.iloc[i,8]= 'bad'
    i+=1
```

### Question 6
